```python
#  Write a Python program to display the current axis limits values and set new axis values. 
import matplotlib.pyplot as plt
x = [1,4,5,6,8]
y = [10,50,100,150,200]

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Axis Check value')
print(plt.axis())
plt.axis([0,10,0,200])
plt.plot(x, y)
```
