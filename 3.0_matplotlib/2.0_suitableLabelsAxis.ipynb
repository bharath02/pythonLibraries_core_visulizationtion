```python
"""Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis and a title"""

import matplotlib.pyplot as plt

X=['A','B','C',"D","E","F"]
Y=[120,80,60,50,40,10]
plt.plot(X,Y)
plt.xlabel("Grades of X Class")
plt.ylabel("Number of Students achived")
plt.title("Line Graph of a X class Grade ")
```
