```python
""" Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file. 
Sample data: 
medal.csv
country,gold_medal
United States,46
Great Britain,27
China,26
Russia,19
Germany,17
"""

import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("medal.csv",sep=",")
print(data)
country=data['country']
goldMedal=data['gold_medal']
plt.pie(goldMedal,explode=(0.1,0,0,0,0,0.1),  labels=country, autopct="%1.1f%%")
```
