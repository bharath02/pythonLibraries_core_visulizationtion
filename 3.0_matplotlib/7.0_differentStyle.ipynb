```python
"""Write a Python program to plot two or more lines with different styles"""
import matplotlib.pyplot as plt

x1=[1,2,3]
y1=[10,20,5]
x2=[1,2,3]
y2=[40,10,5]
plt.plot(x1,y1,color='blue',label='solid',linestyle='solid')
plt.plot(x2,y2,color='green',label='dashdot',linestyle='dashdot')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Two line Plot")
```
