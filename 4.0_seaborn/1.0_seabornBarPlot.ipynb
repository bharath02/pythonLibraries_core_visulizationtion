```python
"""Seaborn Bar plot:
Write a program to draw bar plot of sex against survived for a dataset given in the url

URL : https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv
"""

import pandas as pd
import seaborn as sb
import io
import requests
import matplotlib.pyplot as plt

url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
data=requests.get(url).content
data=pd.read_csv(io.StringIO(data.decode('utf-8')))

sb.barplot(x='sex',y='survived',data=data)
plt.show()
print(sep="   ")
sb.barplot(x='sex',y='survived',hue='class',data=data)
plt.show()
```
